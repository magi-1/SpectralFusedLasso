```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from spectral_ts.models import VARMA
from spectral_ts.estimates import fourier
```

# VARMA

```python
p,q = 20, 20
max_eigenval = .5
sparsity = 0
num_series = 3
N = 200

model = VARMA(p, q, num_series, N, sparsity, max_eigenval)
model
```

```python
model.simulate()
```

```python
for col in model.data:
    plt.figure(figsize = (10,3))
    plt.plot(list(range(model.N)), model.data[col], color = 'k')
    plt.show()
```

# Spectral Density Estimate

```python
m = 2 # smoothing paramater
F = fourier(model.data, m)
```

```python
F.plot_pair(0,0)
F.plot_pair(1,1)
F.plot_pair(0,2)
```
